```python
import csv
import matplotlib.pyplot as plt
top_k = 20
lam = 0.1

def convert_strings_to_floats(input_array):
    output_array = []
    for element in input_array:
        converted_float = float(element)
        output_array.append(converted_float)
    return output_array


fedavg_acc = []
ibcs_acc = []

fedavg_loss = []
ibcs_loss = []

fedavg_power = []
ibcs_power = []
f = open(f'./cifar_results/fedavg_acc_lam_{lam}_k_{top_k}.txt', 'r')
for line in f:
    fedavg_acc.append(line.strip())
fedavg_acc = convert_strings_to_floats(fedavg_acc)

f = open(f'./cifar_results/fedavg_loss_lam_{lam}_k_{top_k}.txt', 'r')
for line in f:
    fedavg_loss.append(line.strip())
fedavg_loss = convert_strings_to_floats(fedavg_loss)

f = open(f'./cifar_results/fedavg_power_lam_{lam}_k_{top_k}.txt', 'r')
for line in f:
    fedavg_power.append(line.strip())
fedavg_power = convert_strings_to_floats(fedavg_power)

f = open(f'./cifar_results/ibcs_acc_lam_{lam}_k_{top_k}.txt', 'r')
for line in f:
    ibcs_acc.append(line.strip())
ibcs_acc = convert_strings_to_floats(ibcs_acc)

f = open(f'./cifar_results/ibcs_loss_lam_{lam}_k_{top_k}.txt', 'r')
for line in f:
    ibcs_loss.append(line.strip())
ibcs_loss = convert_strings_to_floats(ibcs_loss)

f = open(f'./cifar_results/ibcs_power_lam_{lam}_k_{top_k}.txt', 'r')
for line in f:
    ibcs_power.append(line.strip())
ibcs_power = convert_strings_to_floats(ibcs_power)

fig, ax = plt.subplots()
ax.plot(fedavg_acc)
ax.plot(ibcs_acc)

ax.set_title('Accuracy')
ax.legend(['FedAvg', 'IBCS'])
ax.xaxis.set_label_text('Gobal Epochs')
ax.yaxis.set_label_text('Accuracy in %')
plt.savefig(f'./cifar_results/acc_lam{lam}_k_{top_k}.png')
plt.show()


fig, ax = plt.subplots()
ax.plot(fedavg_loss)
ax.plot(ibcs_loss)

ax.set_title('Loss')
ax.legend(['FedAvg', 'IBCS'])
ax.xaxis.set_label_text('Gobal Epochs')
ax.yaxis.set_label_text('Loss')
plt.savefig(f'./cifar_results/loss_lam{lam}_k_{top_k}.png')
plt.show()


fig, ax = plt.subplots()
ax.plot(fedavg_power)
ax.plot(ibcs_power)

ax.set_title('Power')
ax.legend(['FedAvg', 'IBCS'])
ax.xaxis.set_label_text('Gobal Epochs')
ax.yaxis.set_label_text('Power')
plt.savefig(f'./cifar_results/power_lam{lam}_k_{top_k}.png')
plt.show()

```
